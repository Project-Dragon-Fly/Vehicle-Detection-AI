```python
import pandas as pd
def get_route(vehicle,color,**kwarg):
    vehicles = pd.read_excel("data/route_map.xlsx",sheet_name=0)
    
    vehicles['color'] = vehicles.apply(lambda row: row['Vehicle'].split()[0], axis=1)
    vehicles['Vehicle'] = vehicles.apply(lambda row: " ".join(row['Vehicle'].split()[1:]), axis=1)
    vehicles['GPS location'] = vehicles.apply(lambda row: "".join(row['GPS location'].split()), axis=1)
    
    prune = vehicles[vehicles['Vehicle']==vehicle]
    prune = prune[prune['color']==color]
    prune = prune.sort_values('TimeStamp')
    
    result = list()
    for index,row in prune.iterrows():
        result.append((row['Camera'],str(row['TimeStamp']),row['GPS location']))
    return result

get_route('car','red')
```
